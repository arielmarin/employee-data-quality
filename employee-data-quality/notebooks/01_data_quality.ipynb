```python
## Data Cleaning

df_clean = df.copy()
```

```python
df_clean["Join_Date"] = pd.to_datetime(df_clean["Join_Date"])
```

```python
df_clean["Join_Date"].info()
```

```python
df_clean[["Department" , "Region"]] = df_clean["Department_Region"].str.split("-",expand=True)
```

```python
df_clean[["Department", "Region"]].head()
```

```python
df_clean["Department"].isnull().sum()
```

```python
df_clean["Department"].value_counts()
```

```python
df_clean["Region"].isnull().sum()
```

```python
df_clean["Region"].value_counts()
```

```python
df_clean.drop(columns="Department_Region", inplace=True)
```

```python
df_clean.head()
```

```python
df_clean["Phone"] = df_clean["Phone"].astype(str)
```

```python
df_clean["Phone"].info()
```

```python
df_clean.isnull().sum()
```

```python
salary_mean = df_clean["Salary"].mean()

df_clean["Salary"] = df_clean["Salary"].fillna(salary_mean)
```

```python
df_clean["Salary"].isnull().sum()
```

```python
df_clean["Age"].hist(bins=15)

plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.title("Distribución de edades")

plt.show()
```

```python
age_median = df_clean["Age"].median()

df_clean["Age"] = df_clean["Age"].fillna(age_median)
```

```python
df_clean["Age"].isnull().sum()
```

```python
df_clean["Age"] = df_clean["Age"].astype(int)
```

```python
df_clean["Phone"].str.startswith("-").sum()
```

```python
df_clean["Phone"] = df_clean["Phone"].str.lstrip("-")
```
